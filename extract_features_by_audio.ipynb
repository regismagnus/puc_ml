```python
'''Need install
conda install -c conda-forge ffmpeg
'''
from os import listdir
import os.path
import numpy as np
import pandas as pd

#metodos para extração de caracteristicas
from extract_features import *

folder_audio = "./audio"

#where is the tsv
folder_info = "./database"
#file to convert audio
data_audio = "train"
```

```python
#carrega dados tsv
info_data = pd.read_csv(folder_info + "/" + data_audio + ".tsv", sep="\t")
count_client = info_data.groupby("client_id").size().reset_index(name='counts')

#somente usar clientes onde tenha mais do que 10 audios
info_data = info_data[info_data["client_id"].isin(count_client[count_client["counts"] > 10]["client_id"])]
info_data.shape
```

```python
info_data.columns
```

```python
#cria base de dados
features_info = {
    "zcr": 1, "chroma_cqt": 12, "chroma_cens": 12, "tonnetz": 6, "chroma_stft": 12, "rmse": 1, "spectral_centroid": 1,
    "spectral_bandwidth": 1, "spectral_contrast": 7, "spectral_rolloff": 1, "mfcc": 20
}

statistic_col = ["mean", "std", "median", "min", "max"]

if os.path.isfile("./data_" + data_audio + ".csv"):
    #caso processo de extração já tenha sido iniciado carrega dados
    data = pd.read_csv("./data_" + data_audio + ".csv", index_col=0, header=[0, 1, 2])
        
    #coloca na lista de audios somente os que ainda não foram processados
    listTrack = list(data[data['client_id', 'null', 'null'].isnull()].index)
else:
    #lista de audios para processar. Utiliza o nome do arquivo como index
    listTrack = list(info_data.path)
    
    #create empty data base
    columns = [('client_id', "null", "null")]
    for feature, size in features_info.items():
        for stat in statistic_col:
            for i in range(size):
                columns.append((feature, stat, i + 1))

    empty_data = np.zeros((len(listTrack), len(columns)))
                
    data = pd.DataFrame(index=listTrack, columns=pd.MultiIndex.from_tuples(columns))
    data.index.name = "track_id"
```

```python
#salvar a cada 100 audios
count_track = 0

#preenche os dados para todos os arquivos mp3
for track in listTrack:
    #verifica se o arquivo existe
    if not os.path.isfile(folder_audio + "/" + track):
        print("%s Not exist" % track)
        continue
    count_track += 1
    print("Proccess %s number %d" % (track, count_track))
        
    #seta a identificação do cliente na linha do audio
    data.loc[track, ('client_id', 'null', 'null')] = str(info_data[info_data['path'] == track].client_id.values[0])

    #chama metodo para extrair caracteristicas da música e setar na base de dados
    extract_features(data, track, folder_audio + "/" + track)
    
    #salva a cada 100 audios
    if count_track%100 == 0:
        data.to_csv("./data_" + data_audio + ".csv")
```

```python
#salva somente linhas processadas
data[~data['client_id', 'null', 'null'].isnull()].to_csv("./data_" + data_audio + ".csv")
```
